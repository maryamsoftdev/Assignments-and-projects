```python
class LinearRegression:
    def __init__(self):
        self.weights = None

    def train(self, X, Y, alpha=0.0001, max_iterations=1000, print_interval=None):
        # Initialize weights
        self.weights = np.zeros(X.shape[1] + 1)  # including bias term
        num_examples = len(X)
        
        for iteration in range(max_iterations):
            predictions = self.hypothesis(X)
            error = predictions - Y.flatten()  # Flatten Y to match shape of predictions
            
            self.weights[0] -= alpha * np.mean(error)
            self.weights[1:] -= alpha * np.mean(error * X.T, axis=1)

            total_error = self.cost_function(Y, predictions)

            if print_interval and (iteration + 1) % print_interval == 0:
                print(f"Iteration {iteration + 1}, Loss: {total_error}")

            if total_error == 0:
                print(f"Converged at iteration {iteration + 1}")
                break

    def hypothesis(self, X):
        return np.dot(X, self.weights[1:]) + self.weights[0]

    def predict(self, X):
        return self.hypothesis(X)

    def get_weights(self):
        return self.weights[1:]  # exclude bias term

    def cost_function(self, Y, predictions):
        return np.mean((Y - predictions) ** 2)
```

```python
# Load the dataset
df = pd.read_csv("Boston-house-price-data.csv", sep=",")
```

```python
# Check for missing values
if df.isnull().values.any():
    print("Warning: Dataset contains missing values. Please handle them appropriately.")
    # Handle missing values if needed
```

```python
# Normalize the features
X = (df.drop(columns=['MEDV']) - df.drop(columns=['MEDV']).mean()) / df.drop(columns=['MEDV']).std()
X = X.values
```

```python
# Add bias term to X
X = np.hstack((np.ones((X.shape[0], 1)), X))
```

```python
# Extract target variable
Y = df['MEDV'].values.reshape(-1, 1)  # reshaping to a column vector
```

```python
# Create and train the model
model = LinearRegression()
model.train(X, Y, alpha=0.0001, max_iterations=10000, print_interval=1000)
```

```python
# Get the trained weights
print("Trained Weights:", model.get_weights())
```

```python
# Predict
predictions = model.predict(X)
print("Predictions:", predictions)
```
