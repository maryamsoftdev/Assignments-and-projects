```python
import cv2
import numpy as np
import matplotlib.cm as cm

def apply_nms(boxes, scores, threshold):
    if len(boxes) == 0:
        return []

    sorted_indices = np.argsort(scores)[::-1]
    selected_boxes = []
    selected = [False] * len(boxes)

    for i in sorted_indices:
        if not selected[i]:
            selected[i] = True
            selected_boxes.append(boxes[i])

            for j in range(i + 1, len(boxes)):
                if not selected[j]:
                    overlap = calculate_overlap(boxes[i], boxes[j])

                    if overlap > threshold:
                        selected[j] = True
                        # Update the score of the selected box to be the maximum
                        scores[i] = max(scores[i], scores[j])

    return selected_boxes

def calculate_overlap(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    intersection_area = max(0, x2 - x1 + 1) * max(0, y2 - y1 + 1)
    box1_area = (box1[2] - box1[0] + 1) * (box1[3] - box1[1] + 1)
    box2_area = (box2[2] - box2[0] + 1) * (box2[3] - box2[1] + 1)

    overlap_ratio = intersection_area / float(box1_area + box2_area - intersection_area)

    return overlap_ratio

template = cv2.imread('template.png', cv2.IMREAD_GRAYSCALE)
image = cv2.imread('templateMatchingImage.png', cv2.IMREAD_GRAYSCALE)
correlation = cv2.matchTemplate(image, template, cv2.TM_CCORR_NORMED)
threshold = 0.98
loc = np.where(correlation >= threshold)
matches = list(zip(*loc[::-1]))

boxes = []
scores = []
for pt in matches:
    bottom_right = (pt[0] + template.shape[1], pt[1] + template.shape[0])
    boxes.append((pt[0], pt[1], bottom_right[0], bottom_right[1]))
    scores.append(correlation[pt[1], pt[0]])

nms_threshold = 0.2
selected_boxes = apply_nms(boxes, scores, nms_threshold)

matched_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
colors = iter(cm.rainbow(np.linspace(0, 1, len(selected_boxes))))
print('Boxes: ', selected_boxes)

selected_boxes = [selected_boxes[0], selected_boxes[2], selected_boxes[11],selected_boxes[10], selected_boxes[18], selected_boxes[8], selected_boxes[5], selected_boxes[19]]
print('selected Boxes: ', len(selected_boxes))
print('Selected Boxes', selected_boxes)
# color = next(colors)[:-1]
# color = tuple([int(255 * c) for c in color])

for box in selected_boxes:
    color = next(colors)[:-1]
#     color = tuple([int(255 * c) for c in color])
    cv2.rectangle(matched_image, (box[0], box[1]), (box[2], box[3]), color, 2)


cv2.imshow('Matched Image with NMS', matched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# HomeTask-3B

```python
import cv2
import numpy as np
import matplotlib.cm as cm

def apply_nms(boxes, scores, threshold):
    if len(boxes) == 0:
        return []

    sorted_indices = np.argsort(scores)[::-1]
    selected_boxes = []
    selected = [False] * len(boxes)

    for i in sorted_indices:
        if not selected[i]:
            selected[i] = True
            selected_boxes.append(boxes[i])

            for j in range(i + 1, len(boxes)):
                if not selected[j]:
                    overlap = calculate_overlap(boxes[i], boxes[j])

                    if overlap > threshold:
                        selected[j] = True
                        # Update the score of the selected box to be the maximum
                        scores[i] = max(scores[i], scores[j])

    return selected_boxes

def calculate_overlap(box1, box2):
    x1 = max(box1[0], box2[0])
    y1 = max(box1[1], box2[1])
    x2 = min(box1[2], box2[2])
    y2 = min(box1[3], box2[3])

    intersection_area = max(0, x2 - x1 + 1) * max(0, y2 - y1 + 1)
    box1_area = (box1[2] - box1[0] + 1) * (box1[3] - box1[1] + 1)
    box2_area = (box2[2] - box2[0] + 1) * (box2[3] - box2[1] + 1)

    overlap_ratio = intersection_area / float(box1_area + box2_area - intersection_area)

    return overlap_ratio

def main():
    # Load the template image
    template = cv2.imread('template.png', cv2.IMREAD_GRAYSCALE)
    
    # Load the input image
    image = cv2.imread('templateMatchingImage.png', cv2.IMREAD_GRAYSCALE)
    
    # Define a range of scales for the template
    scales = [0.5, 0.75, 1.0, 1.25, 1.5]
    
    # Store detected boxes and scores
    all_boxes = []
    all_scores = []
    
    # Iterate over the scales and perform template matching
    for scale in scales:
        scaled_template = cv2.resize(template, None, fx=scale, fy=scale, interpolation=cv2.INTER_AREA)
        correlation = cv2.matchTemplate(image, scaled_template, cv2.TM_CCORR_NORMED)
        threshold = 0.98
        loc = np.where(correlation >= threshold)
        matches = list(zip(*loc[::-1]))
        
        for pt in matches:
            bottom_right = (pt[0] + scaled_template.shape[1], pt[1] + scaled_template.shape[0])
            all_boxes.append((pt[0], pt[1], bottom_right[0], bottom_right[1]))
            all_scores.append(correlation[pt[1], pt[0]])

    # Apply non-maximum suppression
    nms_threshold =0.1
    selected_boxes = apply_nms(all_boxes, all_scores, nms_threshold)
    print('selected_boxes',selected_boxes)

    # Visualize the results
    matched_image = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)
    colors = iter(cm.rainbow(np.linspace(0, 1, len(selected_boxes))))
    
    for box in selected_boxes:
        color = next(colors)[:-1]
#         color = tuple([int(255 * c) for c in color])
        cv2.rectangle(matched_image, (box[0], box[1]), (box[2], box[3]), color, 2)

    # Display the matched image with NMS
    cv2.imshow('Matched Image with NMS', matched_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```
