How to acess raw data in corpus

```python
r_bible = gutenberg.raw('bible-kjv.txt')
r_bible[:100]
```

```python
r_bible = gutenberg.raw('blake-poems.txt')
r_bible[:100]
```

# How to acess specific words in **corpus**

```python
l_w_bible = gutenberg.words('bible-kjv.txt')
l_w_bible[:10]
```

```python
l_w_bible = gutenberg.words('shakespeare-caesar.txt')
l_w_bible[:10]
```

# How to acess specific sentences in corpus

```python
l_s_bible = gutenberg.sents('shakespeare-caesar.txt')
l_s_bible
```
